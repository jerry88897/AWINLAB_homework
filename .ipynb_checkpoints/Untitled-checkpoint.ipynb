```python
import os
import cv2
import random
import shutil

# 設定要處理的來源資料夾和目的資料夾
source_dir = "./flowers/"
destination_dir1 = "./validation"
destination_dir2 = "./train"

# 獲取來源資料夾下的所有子資料夾
subdirectories = os.listdir(source_dir)

# 從每個子資料夾中隨機選擇 20% 的圖片檔案複製到目的資料夾
for subdirectory in subdirectories:
    # 獲取子資料夾下所有圖片檔案
    files = [x for x in os.listdir(source_dir+"/"+subdirectory) if x.endswith(".jpg")]

    # 計算需要選擇多少個檔案
    num_files_to_select = int(len(files) * 0.1)

    # 隨機選擇指定數量的檔案
    selected_files = random.sample(files, num_files_to_select)

    # 建立目的資料夾
    destination_subdirectory1 = os.path.join(destination_dir1, subdirectory)
    destination_subdirectory2 = os.path.join(destination_dir2, subdirectory)
    
    if not os.path.exists(destination_subdirectory1):
        os.makedirs(destination_subdirectory1)
    if not os.path.exists(destination_subdirectory2):
        os.makedirs(destination_subdirectory2)

    # 將選擇的檔案複製到目的資料夾
    for file in selected_files:
        source_file = os.path.join(source_dir,subdirectory, file)
        destination_file = os.path.join(destination_subdirectory1, file)
        #shutil.copy(source_file, destination_file)
        img = cv2.imread(source_file)
        img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_AREA)
        #img = img.reshape((1,) + img.shape)
        cv2.imwrite(destination_file, img)

    # 將未選擇的檔案複製到目的資料夾
    for file in files:
        if file not in selected_files:
            source_file = os.path.join(source_dir,subdirectory, file)
            destination_file = os.path.join(destination_subdirectory2, file)
            #shutil.copy(source_file, destination_file)
            img = cv2.imread(source_file)
            img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_AREA)
            cv2.imwrite(destination_file, img)
```

```python
import os
import shutil
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=360,  # 角度值，0~360，影象旋轉
    #width_shift_range=0.2,  # 水平平移，相對總寬度的比例
    #height_shift_range=0.2,  # 垂直平移，相對總高度的比例
    shear_range=20,  # 隨機錯切換角度
    #zoom_range=[1, 1.5],  # 隨機縮放範圍
    horizontal_flip=True,  # 一半影象水平翻轉
    vertical_flip=True,
    fill_mode='nearest',  # 填充新建立畫素的方法
    brightness_range=[0.5, 1.3],  # 亮度調整範圍
    channel_shift_range=50,  # 通道偏移
)

train_dir = "train/"
validation_dir = "validation/"

train_generator = datagen.flow_from_directory(
    train_dir,                  # 训练集路径
    target_size=(224, 224),     # 训练集样本尺寸大小为（224， 224）
    batch_size=32,              # 训练集每批包含32个样本
    class_mode='categorical')   
validation_generator = datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical')
```

```python
import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten,Dropout
from keras.preprocessing.image import ImageDataGenerator
import numpy as np

# 使用序贯式模型
model = Sequential()
# 两个3*3*64卷积核 + 一个最大池化层
model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
# 两个3*3*128卷积核 + 一个最大池化层
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
# 三个3*3*56卷积核 + 一个最大池化层
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
# 三个3*3*512卷积核 + 一个最大池化层
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
# 三个3*3*512卷积核 + 一个最大池化层
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
# Flatten层用来将输入“压平”，即把多维的输入一维化，常用在从卷积层到全连接层的过渡。Flatten不影响batch的大小。
# 连接三个全连接层Dense，最后一层用于预测分类。
model.add(Flatten())
model.add(Dense(units=4096,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=4096,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=5, activation="softmax"))

model.summary()
```

```python
# 定义模型优化器， 使用分类交叉熵损失
from keras.optimizers import adam_v2
import tensorflow as tf

lr=1e-4
opt = adam_v2.Adam(learning_rate=lr)
model.compile(optimizer=opt, loss=keras.losses.categorical_crossentropy, metrics=['accuracy', tf.keras.metrics.TopKCategoricalAccuracy(k=5)])
```

```python
# 定义模型和精度计算方式
from keras.callbacks import ModelCheckpoint, EarlyStopping,ReduceLROnPlateau
checkpoint = ModelCheckpoint("vgg16_1.h5", monitor='val_accuracy', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', save_freq='epoch')
early = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=20, verbose=1, mode='auto')
# 設定lr降低條件(0.001 → 0.005 → 0.0025 → 0.00125 → 下限：0.0001)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5,patience=5, mode='min', verbose=1,min_lr=1e-6)
```

```python
hist = model.fit(train_generator,steps_per_epoch=32, validation_data= validation_generator,validation_steps=10,epochs=120,callbacks=[checkpoint, early, reduce_lr])
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=1200)
plt.plot(hist.history["accuracy"])
plt.plot(hist.history['val_accuracy'])
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy","Validation Accuracy","loss","Validation Loss"])
plt.show()
```
